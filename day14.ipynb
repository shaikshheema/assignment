```python
'''Assignment 2: Drug Effectiveness and Side Effects Comparison
Objective:
Compare drug effectiveness and side effects by different products, regions, and trial periods. Visualize these comparisons using matplotlib and seaborn.
Instructions:
Load the dataset.
Perform any necessary data cleaning.
Create the following visualizations:
A bar plot comparing the average Effectiveness for each drug across different regions.
A violin plot to show the distribution of Effectiveness and Side_Effects for each product.
A pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend.
A boxplot comparing Effectiveness for different trial periods.
A regression plot to analyze how Marketing_Spend affects drug Effectiveness.
Based on the visualizations, provide an analysis of:
Which product has the best overall effectiveness.
The correlation between effectiveness and side effects.
Expected Outcome:
Insights into how the effectiveness of each drug varies by region and trial period.
Understanding the trade-off between drug effectiveness and side effects.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Day_13_Pharma_data.csv')

# Data Cleaning (handle missing values and check data types)
df.isnull().sum()  # Check for missing values
df = df.dropna()  # Drop missing values (or use df.fillna() if you prefer to fill them)
df.dtypes  # Check data types to ensure they are correct

# 1. Bar plot comparing the average Effectiveness for each drug across different regions
plt.figure(figsize=(10, 6))
sns.barplot(x='Product_Name', y='Effectiveness', hue='Region', data=df)
plt.title('Average Effectiveness by Drug and Region')
plt.xlabel('Product')
plt.ylabel('Average Effectiveness')
plt.show()

# 2. Violin plot to show the distribution of Effectiveness and Side_Effects for each product
# 2. Violin plot to show the distribution of Effectiveness and Side_Effects for each product
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, inner="quart")
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, inner="quart", color='lightgray')
plt.title('Distribution of Effectiveness and Side Effects for each Product')
plt.xlabel('Product')
plt.ylabel('Value')
plt.show()


# 3. Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# 4. Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(10, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness Distribution by Trial Period')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# 5. Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(10, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Effectiveness vs Marketing Spend')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()

# Analysis: Which product has the best overall effectiveness?
product_effectiveness = df.groupby('Product_Name')['Effectiveness'].mean().sort_values(ascending=False)
print("Average Effectiveness by Product:")
print(product_effectiveness)


# Analysis: The correlation between effectiveness and side effects
effectiveness_side_effects_corr = df[['Effectiveness', 'Side_Effects']].corr()
print("Correlation between Effectiveness and Side Effects:")
print(effectiveness_side_effects_corr)
```
