```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace with actual file path)
df = pd.read_csv('Day_9_banking_data.csv')

# 1. Plot the total sum of Transaction_Amount per Account_Type using a bar plot
transaction_sum_per_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
transaction_sum_per_account_type.plot(kind='bar', color='skyblue')
plt.title('Total Sum of Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Create a pie chart to show the percentage of transactions per Branch
transaction_count_per_branch = df['Branch'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
transaction_count_per_branch.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], startangle=90)
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # To remove the ylabel which is not needed in a pie chart
plt.tight_layout()
plt.show()
```
